# duplicated()
## Checks whether the next occurences are present for any particular value

```python
df = pd.read_csv("csvfiles/employees.csv" , parse_dates=["Start Date" , "Last Login Time"])
df["Gender"] = df["Gender"].astype("category")
df["Senior Management"] = df["Senior Management"].astype("category")
df["Team"] = df["Team"].astype("category")
df.info()
```

```python
df.sort_values("First Name" , inplace=True)
df.head()
```

### In this way we can get the duplicate rows in a dataframe

```python
df[df["First Name"].duplicated()].head()
```

### In this way we can get the duplicate rows in a dataframe starting from end

```python
df[df["First Name"].duplicated(keep="last")].head()
```

### Remove all rows that occured more than once

```python
df[df["First Name"].duplicated(keep=False)].head()
```

### Remove all unique

```python
df[~df["First Name"].duplicated(keep=False)].head()
```

# drop_duplicates()

```python
df.head()
```

### Keep the first occurence of First name

```python
df.drop_duplicates("First Name" , keep="first").head()
```

### Remove all rows for which there are duplicates 

```python
df.drop_duplicates("First Name" , keep=False).head()
```

### Remove if both first name and team are same

```python
df.drop_duplicates(subset=["First Name" , "Team"] , keep=False).head()
```
